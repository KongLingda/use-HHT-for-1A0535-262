```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
colors = {
    'red': '#E24A33',
    'blue': '#348ABD',
    'green': '#8EBA42',
    'orange': '#FBC15E',
    'purple': '#988ED5',
    'pink': '#FFB5B8',
    'gray': '#777777'
}
```

```python
#read txt files
df0 = pd.read_csv('/Volumes/WD_BLACK/Data/1A0535p262/HXMT/date_combine/20201120/avg.qdp',sep=' ')
df01 = pd.read_csv('/Volumes/WD_BLACK/Data/1A0535p262/HXMT/date_combine/20201208/avg.qdp',sep=' ')
df02 = pd.read_csv('/Volumes/WD_BLACK/Data/1A0535p262/HXMT/date_combine/20201213/avg.qdp',sep=' ')

df1 = pd.read_csv('20201120_QPOphase/0001.qdp', sep=' ') 
df11 = pd.read_csv('20201208_QPOphase/0001.qdp', sep=' ') 
df12 = pd.read_csv('20201213_QPOphase/0001.qdp', sep=' ') 

df2 = pd.read_csv('20201120_QPOphase/0506.qdp', sep=' ') 
df21 = pd.read_csv('20201208_QPOphase/0506.qdp', sep=' ') 
df22 = pd.read_csv('20201213_QPOphase/0506.qdp', sep=' ') 

df3 = pd.read_csv('20201120_QPOphase/0001_del.qdp', sep=' ') 
df31 = pd.read_csv('20201208_QPOphase/0001_del.qdp', sep=' ') 
df32 = pd.read_csv('20201213_QPOphase/0001_del.qdp', sep=' ') 

df4 = pd.read_csv('20201120_QPOphase/0506_del.qdp', sep=' ') 
df41 = pd.read_csv('20201208_QPOphase/0506_del.qdp', sep=' ') 
df42 = pd.read_csv('20201213_QPOphase/0506_del.qdp', sep=' ') 
```

```python
#for Average Spectru
#read energies
x0,x01,x02=np.array(df0['E']),np.array(df01['E']),np.array(df02['E'])
x0_err,x01_err, x02_err=np.array(df0['E_err']),np.array(df01['E_err']),np.array(df02['E_err'])
#read spectra
y0,y01,y02=np.array(df0['Flux']),np.array(df01['Flux']),np.array(df02['Flux'])
y0_err,y01_err,y02_err=np.array(df0['Flux_err']),np.array(df01['Flux_err']),np.array(df02['Flux_err'])
#read models
y0_add1,y01_add1,y02_add1=np.array(df0['Add1']),np.array(df01['Add1']),np.array(df02['Add1'])
y0_add2,y01_add2,y02_add2=np.array(df0['Add2']),np.array(df01['Add2']),np.array(df02['Add2'])

#for 0001 Spectru
#read energies
x1,x11,x12=np.array(df1['E']),np.array(df11['E']),np.array(df12['E'])
x1_err,x11_err, x12_err=np.array(df1['E_err']),np.array(df11['E_err']),np.array(df12['E_err'])
#read spectra
y1,y11,y12=np.array(df1['Flux']),np.array(df11['Flux']),np.array(df12['Flux'])
y1_err,y11_err,y12_err=np.array(df1['Flux_err']),np.array(df11['Flux_err']),np.array(df12['Flux_err'])
#read models
y1_add1,y11_add1,y12_add1=np.array(df1['Add1']),np.array(df11['Add1']),np.array(df12['Add1'])
y1_add2,y11_add2,y12_add2=np.array(df1['Add2']),np.array(df11['Add2']),np.array(df12['Add2'])

#for 0506 Spectru
#read energies
x2,x21,x22=np.array(df2['E']),np.array(df21['E']),np.array(df22['E'])
x2_err,x21_err,x22_err=np.array(df2['E_err']),np.array(df21['E_err']),np.array(df22['E_err'])
#read spectra
y2,y21,y22=np.array(df2['Flux']),np.array(df21['Flux']),np.array(df22['Flux'])
y2_err,y21_err,y22_err=np.array(df2['Flux_err']),np.array(df21['Flux_err']),np.array(df22['Flux_err'])
#read models
y2_add1,y21_add1,y22_add1=np.array(df2['Add1']),np.array(df21['Add1']),np.array(df22['Add1'])
y2_add2,y21_add2,y22_add2=np.array(df2['Add2']),np.array(df21['Add2']),np.array(df22['Add2'])

#for 0001 residuals
x3,x31,x32=np.array(df3['E']),np.array(df31['E']),np.array(df32['E'])
x3_err,x31_err, x32_err=np.array(df3['E_err']),np.array(df31['E_err']),np.array(df32['E_err'])
y3,y31,y32=np.array(df3['Delta']),np.array(df31['Delta']),np.array(df32['Delta'])
y3_err,y31_err,y32_err=np.array(df3['Delta_err']),np.array(df31['Delta_err']),np.array(df32['Delta_err'])

#for 0506 residuals
x4,x41,x42=np.array(df4['E']),np.array(df41['E']),np.array(df42['E'])
x4_err,x41_err, x42_err=np.array(df4['E_err']),np.array(df41['E_err']),np.array(df42['E_err'])
y4,y41,y42=np.array(df4['Delta']),np.array(df41['Delta']),np.array(df42['Delta'])
y4_err,y41_err,y42_err=np.array(df4['Delta_err']),np.array(df41['Delta_err']),np.array(df42['Delta_err'])
```
